# Named Entity Recognition 

```python
import facebook_scraper
import nltk
import spacy
```

```python
import flair
from flair.data import Sentence
from flair.models import SequenceTagger

tagger = SequenceTagger.load("flair/ner-dutch-large")
```

```python
def ner_replace_names(text):
    sentence_text = Sentence(text)
    tagger.predict(sentence_text)
    
    per_entities = []
    for entity in sentence_text.get_spans('ner'):
        if entity.tag == "PER":
            per_entities.append(entity.text)
    
    for entity in per_entities:
        text = text.replace(entity, "ANONYMOUS")
        
    return text
```

```python
string_tex = "Hoi dit is Lotte de Groot. Hoe gaat het met jou Luca Pamboer? Ik ben aan het wachten op Ymke Knoben-Zeideveld en Anne-Marie van Vleuten"
print(ner_replace_names(string_tex))
```

# Facebook data scraping

Public "Lactosevrij" --> 1724293977830000 (https://www.facebook.com/groups/1724293977830000)

```python
from facebook_scraper import get_posts
from nltk import pos_tag, word_tokenize
import re

post_count = 0
all_posts = []
full_posts = []
post_texts = []
post_comments = []
post_text_comment = []

for post in get_posts(group = '1724293977830000', credentials=('lottedegroot123@gmail.com' , 'nOBW9%6P@'), options={"comments": True}):
    if post_count < 500 and post not in full_posts:
        specific_post_text_comment = ""
        
        post_text = post['post_text']
        anon_text = ner_replace_names(post_text)
        post_texts.append(anon_text)
        
        specific_post_text_comment += "POST" + str(post_count) + ": " + anon_text + "\n"
        specific_post_comments = ""
        comment_count = 0
        for i in post['comments_full']:
            specific_post_comments += "COM" + str(comment_count) + ": " + i['comment_text'] + "\n"
            reply_count = 0
            if len(i['replies']) != 0:
                for j in i['replies']:
                    specific_post_comments += "REP" + str(reply_count) + "COM" + str(comment_count) + ": " + j['comment_text'] + "\n"
                    reply_count += 1
            comment_count += 1
        specific_post_comments += "END"
        
        anon_specific_post_comments = ner_replace_names(specific_post_comments)   
        post_comments.append(anon_specific_post_comments)
        
        specific_post_text_comment += anon_specific_post_comments
        post_text_comment.append(specific_post_text_comment)
        
        post_count += 1
        
    elif post in full_posts:
        print("POST THAT HAS NOT BEEN ADDED ", post)
        index_duplicate = full_posts.index(post)
        print("DUPLICATE VERSION ON INDEX ", index_duplicate, ": \n", post[index_duplicate])
```

```python
import emoji

def createFilesOfPosts(posts, folder, fileSpecification):
    for i in range(0, len(posts)):
        with open("C:/Users/Lotte/OneDrive/AI M1/Text and Multimedia Mining/TxMM project/" + folder + "/" + str(i) + fileSpecification + ".txt", "w", encoding = 'utf-8') as file:
            print(i)
            post = posts[i]
            file.write(emoji.demojize(str(post))) #demojize turns emoji characters into text, e.g. a bomb emoji into ":bomb:"
```
