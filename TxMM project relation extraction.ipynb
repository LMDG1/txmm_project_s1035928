# Regexes:

3 types of relations: 
   1. A **mention** of intolerance (in the LI forum) --> less likely to be a relevant relation
   2. (intolerance, intolerance) as personal **experience** --> likely to be a relevant relation
   3. (symptom, intolerance) as a **suggestion/advice** --> highly likely to be a relevant relation
   
The three are ordered from lowest - highest relevance

## Low relevance (= mentions)
- In close proximity a mention of a symptom + an intolerance
- A description of an intolerance


## Relevant (= experiences)
- In close proximity the mention of lactose / a lactose intolerancy + a different (possible) intolerance
    - Either 'and'/'also'
    - Usually mentioned out of personal experience, a statement about themselves, others or a (self-)suspected combination of two intolerances
    - Can also be more general, as long as an intolerance is tied to lactose (intolerance) in close proximity



## High relevance (= advice)
- A statement of advice (about a possible intolerance/sensitivity) after a mention of symptoms
    - Either post text (symptoms) / comments (advice/suggestion) or...
    - Comment (symptoms) / replies (advice/suggestion
    - Regexes like, "misschien kan je niet goed tegen ..."

```python
import regex as re

#Regex with f-string so that variables are easily identified
#Regexes only done on mentioned intolerances to save computation time
def get_regexes(mentioned_intolerances):
    
    #Low relevance: regexes that look for an intolerance (+ symptom description) mention in the full text of a post
    text_regexes = [
        [re.compile(f"{sym_des}[-\s,]*(.{1,20})(van|door|voor)?[-\s,]*{into}") for sym_des in descriptions for into in mentioned_intolerances],
        [re.compile(f"{sym_des}[-\s,]*(te)?[-\s,]*(veel)[-\s,]*{into}") for sym_des in descriptions for into in mentioned_intolerances],
        [re.compile(f"{into}[-\s,]*{sym_des}") for into in mentioned_intolerances for sym_des in descriptions],
        [re.compile(f"{gen_des}[-\s,]*(te)?[-\s,]*(zijn)[-\s,]*voor[-\s,]*{into}") for gen_des in general_descriptions for into in mentioned_intolerances],
        [re.compile(f"{into}[-\s,]*{gen_des}") for into in mentioned_intolerances for gen_des in general_descriptions],
        [re.compile(f"(buik|darmen|huid|kan)?[-\s,]*{sym_exp}([-\s,]*de)?[-\s,]*{into}([-\s,]*te[-\s,]*kunnen)?") for sym_exp in symptom_expressions for into in mentioned_intolerances],
        [re.compile(f"van(.{1,100}){into}(.{1,100}){sym_des}") for into in mentioned_intolerances for sym_des in symptom_descriptions]
    ]

    #Relevant: regexes that look for a lactose (intolerance) + other intolerance relation in a sentence 
    experience_regexes = [
        [re.compile(f"van[-\s,]+{into}.*last") for into in mentioned_intolerances],
        [re.compile(f"lactose.*[-\s,]+(ook)[-\s,]+.*{into}[-\s,]*{gen_des}") for gen_des in general_descriptions for into in mentioned_intolerances],
        [re.compile(f"lactose[-\s,]*{gen_des}.*[-\s,]+(ook)[-\s,]+.*{into}[-\s,]*{gen_des}?") for gen_des in general_descriptions for into in mentioned_intolerances],
        [re.compile(f"lactose.*[-\s,]+(en)[-\s,]+.*{into}[-\s,]*{gen_des}") for into in mentioned_intolerances for gen_des in general_descriptions],
        [re.compile(f"lactose[-\s,]*{gen_des}.*[-\s,]+(en)[-\s,]+.*{into}[-\s,]*{gen_des}?") for into in mentioned_intolerances for gen_des in general_descriptions],
        [re.compile(f"{into}.*[-\s,]+(en)[-\s,]+.*lactose(-|\s)?{gen_des}?") for into in mentioned_intolerances for gen_des in general_descriptions ],
        [re.compile(f"{into}[-\s,]*{gen_des}.*[-\s,]+en[-\s,]+.*lactose") for into in mentioned_intolerances for gen_des in general_descriptions ],
        [re.compile(f"{sym_des}.*(ver)?denk(ing)?.*{into}") for sym_des in symptom_descriptions for into in mentioned_intolerances],
        [re.compile(f"(ver)denk(ing)?.*{into}.*{sym_des}") for into in mentioned_intolerances for sym_des in symptom_descriptions],
        [re.compile(f"lactose.*[-\s,]+(en|ook)[-\s,]+.*{sym_des}[-\s,]*{into}") for sym_des in symptom_descriptions for into in mentioned_intolerances],
        [re.compile(f"{sym_des}.*lactose[-\s,]*(en|ook)?[-\s,]*{into}") for sym_des in symptom_descriptions for into in mentioned_intolerances],
        [re.compile(f"{sym_des}.*{into}[-\s,]+(en|ook)?[-\s,]*lactose") for sym_des in symptom_descriptions for into in mentioned_intolerances]
    ]

    #High Relevance: 
    
    #Regex looks for symptoms in post (to tie to found advice below)
    post_text_regexes = [re.compile(f"{sym_all}") for sym_all in symptom_all]

    #Regexes look for given advice or suggestions in comments/replies of symptom mentions
    comment_text_regexes = [
        [re.compile(f"{sym_des}[-\s,]*(.{1,20})(van|door|voor)?[-\s,]*{into}\?*") for sym_des in descriptions for into in mentioned_intolerances],
        [re.compile(f"(buik|darmen|huid|kan)?[-\s,]*{sym_exp}([-\s,]+de)?[-\s,]+{into}\?*") for sym_exp in symptom_expressions for into in mentioned_intolerances],
        [re.compile(f"(ge)?test([-\s,]+op)?[-\s,]+{into}\?*") for into in mentioned_intolerances],
        [re.compile(f"zou.*{into}+?{gen_des}?.*kunnen[-\s,]+hebben\?*") for into in mentioned_intolerances for gen_des in general_descriptions],
        [re.compile(f"kan.*{into}[-\s,]+zijn\?*") for into in mentioned_intolerances]
    ]
    
    return text_regexes, experience_regexes, post_text_regexes, comment_text_regexes
```

```python
#Extracts information from a text (can be full text or a single sentence) based on 'regexes'
def extract_relations (text, regexes, mentioned_intolerances):
    relations = []
    if any(into in text for into in mentioned_intolerances):
        for pattern in regexes:
            for p in pattern:
                match = re.search(p, text)
                if match:
                    relations.append(match.group())
    if len(relations) == 2:
        if relations[0] in relations[1]:
            relations.remove(relations[0])
    if len(relations) == 2: 
        if relations[1] in relations[0]:
            relations.remove(relations[1])
    return relations

#Extracts information from two separate texts. Only returns if a relation is found between the texts
def extract_separated_relation (postText, commentText, postTextRegexes, commentTextRegexes):
    separated_relations = []
    symptoms_found = []
    advice_found = []
    for pattern1 in postTextRegexes:
        match1 = re.search(pattern1, postText)
        if match1:
            symptoms_found.append(match1.group())
    for pattern2 in commentTextRegexes:
        for p2 in pattern2:
            match2 = re.search(p2, commentText)
            if match2:
                advice_found.append(match2.group())
    for a in advice_found:
        if len(symptoms_found) > 0 :
            separated_relations.append((symptoms_found, a))    
    return separated_relations
```

```python
import pandas as pd
import nltk

def relations_dataframe(postData, wanted_indexes, re_text, re_exp, re_post, re_comment, mentioned_intos):
    #Create the dataframe that will contain all data
    post_relations = pd.DataFrame(columns = ["Post", "Relations", "Relevance", "Type"])

    #Loop through all post data
    for i in range(0, len(wanted_indexes)):
        actual_index = wanted_indexes[i]
        post = postData[i]
        
        ###############################################################
        #Extract mentions from full text (= post + comment + reply data)
        relations_of_full_text = extract_relations(post.lower(), re_text, mentioned_intos)
        relations_of_full_text = list(dict.fromkeys(relations_of_full_text)) #Removes duplicates
    
    
        ################################################################
        #Extract lactose (intolerance) + other intolerance relations
        #First, sentences are created 
        nsplit = post.split('\n') #First split based on new lines so that all separate comments/replies are split
        sentences = []
        for sens in nsplit:
            new_element = nltk.tokenize.sent_tokenize(sens) #Splits into sentences based on punctuation 
            sentences = sentences + new_element
            
        #Second, relations are extracted from sentences
        sentence_relations = []
        for s in sentences:
            s_relations = extract_relations(s.lower(), re_exp, mentioned_intos)
            s_relations = list(dict.fromkeys(s_relations)) #Removes duplicates
            #Remove relations that are a subset of another one 
            #E.g. extracted 'soja intolerant' is subset ofsoja intolerantie
            if len(s_relations) == 2: 
                if s_relations[0] in s_relations[1]:
                    s_relations.remove(s_relations[0])
            if len(s_relations) == 2: 
                if s_relations[1] in s_relations[0]:
                    s_relations.remove(s_relations[1])
            if len(s_relations) > 0:
                sentence_relations.append(s_relations) #Final list of extracted sentence relations
        
        
        #######################################################################################################
        #Extract relations on split data (extract symptoms from one text and intolerance advice from the other)
        
        #First, needed data/variables are prepared
        advice_relations = []
        post_text = post.split('COM0:', maxsplit=1)[0] #Extracts only the post text out of all the post text   
        comment_indexes = []
        comments = "" #Will collect all comments of a single post in one string
        for i in range(0, len(nsplit)):
            if nsplit[i].startswith("COM"): #Scraped data with indicators such as "COM" to allow this
                comment_indexes.append(i)       
        for i in range(0, len(comment_indexes)):
            comment = nsplit[comment_indexes[i]]
            comments = comments + comment            
            replies = "" #Will collect all replies of a single comment in one string 
            if comment_indexes[i] != comment_indexes[-1]:
                for j in range(comment_indexes[i]+1, comment_indexes[i+1]):
                    if nsplit[j].startswith("REP"):
                        replies = replies + nsplit[j]
                if len(replies) > 0:
                    
                    #Second, Extract symptoms from comments and intolerance advice from comment replies
                    comment_reply_relations = extract_separated_relation(comment.lower(), replies.lower(), re_post, re_comment)
                    if len(comment_reply_relations) > 0:
                        advice_relations.append(comment_reply_relations) 
                        
        #Third, extract symptoms from post text and intolerance advice from direct comments
        post_comment_relations = extract_separated_relation(post_text.lower(), comments.lower(), re_post, re_comment)
        if len(post_comment_relations) > 0:
            advice_relations.append(post_comment_relations)
            
        
        ########################################################################    
        #Check for doubles or subsets of extracted relations to remove duplicate data
        
        #Check for doubles in mentions and advice relations 
        if (len(advice_relations)>0):
            for i in advice_relations:                
                for j in relations_of_full_text:
                    if len(i) > 0 and i[0][1] == j:
                        relations_of_full_text.remove(j)
                    elif len(i) > 0 and j in i[0][1]:
                        relations_of_full_text.remove(j)
                        
        #Check for doubles in mentions and sentence_relations 
        if (len(sentence_relations)>0):
            for i in sentence_relations:                
                for j in relations_of_full_text:
                    for r in i:
                        if r == j:
                            relations_of_full_text.remove(j)
                        elif j in r:
                            relations_of_full_text.remove(j)
                            
                             
        ###################################                
        #Add the extracted relations to the dataframe with correct relevance and type
        
        #Uncertain: (mention) relations on full text
        relations_new_row = pd.DataFrame({"Post":str(actual_index), 
                                          "Relations":relations_of_full_text,
                                          "Relevance": "Less relevant",
                                          "Type": "Mention"})
        post_relations = pd.concat([post_relations, relations_new_row])
        
        #Certain: (experience) relations on sentences
        sentence_relations_new_row = pd.DataFrame({"Post":str(actual_index), 
                                          "Relations":sentence_relations,
                                          "Relevance": "Relevant",
                                          "Type": "Experience"})
        post_relations = pd.concat([post_relations, sentence_relations_new_row])
        
        #Highly certain: (symptom/advice) relations of post/comments and comment/replies
        separated_relations_new_row = pd.DataFrame({"Post":str(actual_index), 
                                          "Relations":advice_relations,
                                          "Relevance":"Highly relevant",
                                          "Type": "Advice"})
        post_relations = pd.concat([post_relations, separated_relations_new_row])
    
    post_relations = post_relations.set_index('Post')
    return post_relations
```
